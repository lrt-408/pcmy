```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#生成时间序列
t = np.arange(-2 * np.pi,2 * np.pi, 0.05)
# sin(2x)
s = np.sin(2 * t)
# cos
c = np.cos(t)
# sin(x)*cos(5x)
s_1 = np.sin(t) * np.cos(5 * t)
# y=2x^2+4x+6
s_2 = 2 * t * t + 4 * t + 6
```

```python
# 绘图
plt.plot(t, s, "r.", label = "sin(x)")
plt.plot(t, c, "*",label = "cos(x)")
plt.legend()
plt.show()
```

```python
plt.figure(1)
ax1 = plt.subplot(4,2,1)#2行2列第1个区
ax2 = plt.subplot(4,2,2)
ax3 = plt.subplot(3,1,2)
ax4 = plt.subplot(4,1,4)
plt.sca(ax1)
plt.plot(t, s, 'b')
plt.sca(ax2)
plt.plot(t, c, 'r')
plt.sca(ax3)
plt.plot(t, s_1, 'y')
plt.sca(ax4)
plt.plot(t, s_2, 'y')
plt.show()
```
